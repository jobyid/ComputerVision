### Import statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cv2.__version__
```

```python
def imshow(img):
    cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.imshow(img)
```

### To edit an image making a copy of it is always a good idea!

```python
# Load an image
# And make a copy of the original image to avoid manipulating the original image
img = cv2.imread('img/m.jpg')
img_copy = img.copy()
img
```

[Hint](# "Numpy .copy function") 

## Drawing

### Drawing lines

```python
# Draw some lines on top of the image 
cv2.line(img_copy, (0,0),(100,500),(255,0,255),5)
imshow(img_copy)
```

[Hint](# "cv2.line function") 

### Circles

```python
# Draw some circles on top of the image
cv2.circle(img_copy,(300,300),55,(255,0,0),3)
imshow(img_copy)
```

[Hint](# "cv2.circle") 

### Ellipses

```python
# Draw some elipses on top of the image
cv2.ellipse(img_copy,(250, 150), (80, 20), 5, 0, 360, (0,255,0), -1)
imshow(img_copy)
```

[Hint](# "cv2.ellipse") 

### Poligons

```python
# Lets draw some poligons, first declare the points of the poligon
pts = np.array([[(12,22),(230,121),(55,78),(23,45)]])
cv2.polylines(img_copy,pts,True, (255,150,255),4)
imshow(img_copy)
```

### Add an overlay on top of an image

```python
# Load an image and create a copy of it you can call it overlay
# Create a rectangle with a fill color (-1) on top of the overlay image
# Combine both images setting some opacity on the overlay
# Add some text on top of the overlay and show the image on screen
i = cv2.imread('img/blue-flowers.jpg')
base = i.copy()
overlay = i.copy()
cv2.rectangle(overlay,(10,10),(170,70), (0,0,0),-1)
cv2.putText(overlay,"Text Example",(40,40), cv2.FONT_HERSHEY_PLAIN, 1, (255,255,255),2)
alpha = 0.5
x = cv2.addWeighted(overlay,alpha,base,1-alpha,0)
imshow(x)
```

[Hint](# "cv2.addWeighted")
