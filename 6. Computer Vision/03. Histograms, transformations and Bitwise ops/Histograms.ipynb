# Computer Vision Histograms

### Import statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

## Histograms

```python
# Load some images, calculate and plot their color histograms, one color at the time
# try it with different color spaces BGR, HSV, etc
blue_flowers   = cv2.imread('img/blue-flowers.jpg')
blue_channels  = cv2.split(blue_flowers) 
blue_f_B_hist  = cv2.calcHist(blue_channels,[0],None,[256],(0,256))
blue_f_G_hist  = cv2.calcHist(blue_channels, [1], None, [256], (0,256))
blue_f_R_hist  = cv2.calcHist(blue_channels, [2], None, [256], (0,256))
purple_flowers = cv2.imread('img/purple-flowers.jpg')
pur_ch         = cv2.split(purple_flowers)
pur_f_B_hist   = cv2.calcHist(pur_ch,[0], None, [256], (0,256))
pur_f_G_hist   = cv2.calcHist(pur_ch,[1], None, [256], (0,256))
pur_f_R_hist   = cv2.calcHist(pur_ch,[2], None, [256], (0,256))
surfer  = cv2.imread('img/surfer.jpg')
sur_ch         = cv2.split(surfer)
sur_f_B_hist   = cv2.calcHist(sur_ch,[0], None, [256], (0,256))
sur_f_G_hist   = cv2.calcHist(sur_ch,[1], None, [256], (0,256))
sur_f_R_hist   = cv2.calcHist(sur_ch,[2], None, [256], (0,256))
```

```python
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(15,15))

fig.tight_layout
axs[0,0].plot(blue_f_B_hist)
axs[0,1].plot(blue_f_G_hist, color="green")
axs[0,2].plot(blue_f_R_hist, color="red")
axs[1,0].plot(pur_f_B_hist)
axs[1,1].plot(pur_f_G_hist, color="green")
axs[1,2].plot(pur_f_R_hist, color="red")
axs[2,0].plot(sur_f_B_hist)
axs[2,1].plot(sur_f_G_hist, color="green")
axs[2,2].plot(sur_f_R_hist, color="red")
plt.show()
```
