```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Template Matching

```python
# Load img/WandoScene.jpg  as the main image
scene = cv2.imread('img/WaldoScene1.jpg')
# Load img/waldo.jpg as the template
wald = cv2.imread('img/waldo.jpg')

# Corvert the images to grayscale
gray_scene = cv2.cvtColor(scene, cv2.COLOR_BGR2GRAY)
gray_wald  = cv2.cvtColor(wald, cv2.COLOR_BGR2GRAY)
w, h = (50,50)
# Find the template in the image
m = cv2.matchTemplate(gray_scene,gray_wald,cv2.TM_CCOEFF)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(m)
# Draw a rectagle around the the detected match
cv2.rectangle(scene,max_loc,(max_loc[0] + w,max_loc[1]+h), (0,255,0),2)
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(scene, cv2.COLOR_BGR2RGB))
```
