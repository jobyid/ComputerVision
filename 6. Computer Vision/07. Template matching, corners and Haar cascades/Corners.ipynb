```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Corners

```python
image = cv2.imread('img/chess.jpg')
plt.imshow(image);
```

```python
# We will work with the image above
# Convert it to grayscale
gray_chess = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect the corners using the cornerHarris
corners = cv2.cornerHarris(gray_chess,3,3,1)

# Display the image with the detected corners 

plt.imshow(corners)
```
